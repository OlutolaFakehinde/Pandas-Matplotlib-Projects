```python
# linestyle - 'dotted', 'dashed'
# line colour c - 'red', 'blue' etc
# linewidth - 10

# create a line chart showing the daily trend of sales based on total amount made

# first group by date because is appearing multiple times [we need to aggregate our total together for each day]

daily_sales = sales_df.groupby(['date'])[['total']].sum()
daily_sales

# 'date' is no the 'index' and 'total' the value
```

```python
plt.plot(daily_sales)
plt.show()
```

```python
plt.plot(daily_sales)
plt.show()
```

## Scatter Plot

```python
quantity = sales_df['quantity'].head(100)
total = sales_df['total'].head(100)

# create a scatter plot
plt.scatter(x = quantity, y = total)

# or color = 'red', s means size, alpha is transparency

# adding title and labels

plt.show()
```

```python
quantity = sales_df['quantity'].head(100)
total = sales_df['total'].head(100)

# create a scatter plot
plt.scatter(x = quantity, y = total, c = 'red', s = 20, alpha = 0.5)

# or color = 'red', s means size, alpha is transparency

# adding title and labels
plt.title('Scattereplot of Quantity and Total')
plt.xlabel('Quantity')
plt.ylabel('Total')
plt.grid(axis = 'y', linewidth = 1)

plt.show()
```

## Subplots

```python
# allows two plots together

plt.subplot(1, 2, 1) # (1, 2, 1) means 1-single row, 2-columns, position 1
plt.hist(sales_df['quantity'], color = 'red')

plt.subplot(1, 2, 2)
plt.hist(sales_df['total'], color = 'green')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10, 7))

ax1.hist(sales_df['quantity'], color = 'blue')
ax2.hist(sales_df['total'], color = 'brown')

plt.show()
```
